```python
import pandas as pd 
```

```python
path_train = "D:\\datasets\\mobile_price_classification\\train.csv"
file = pd.read_csv(path_train)
df = file.copy()
df.sample(10)
```

```python
df.info()
```

```python
df.price_range.describe()
```

```python
label = df.price_range
features = df.drop(['price_range'],axis=1)
```

```python
Corr = features.corr()
for col in Corr.columns :
    print(Corr[f'{col}'].sort_values(ascending=False))
    print('-------------------------------------------')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x = pd.DataFrame(scaler.fit_transform(features),columns=features.columns)
scaled_x.head()
```

```python
path_test = "D:\\datasets\\mobile_price_classification\\test.csv"
file_test = pd.read_csv(path_test,index_col=0)
df_test = file_test.copy()
df_test.head()
```

```python
df_test.info()
```

```python
scaled_test = pd.DataFrame(scaler.fit_transform(df_test),columns=df_test.columns)
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(scaled_x,label,test_size=0.01,random_state=32)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(xtr,ytr)
```

```python
model.score(xtr,ytr)*100
```

```python
model.score(xts,yts)*100
```

```python
model.score(scaled_x,label)*100
```

```python
preds = model.predict(scaled_test)
```
